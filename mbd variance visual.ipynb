```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mbd = pd.read_csv("mbd_train.csv")
mbd.head()
```

```python
mbd = mbd.rename(columns={'microbusiness_density':'mbd'})
```

```python
mbd['pop'] = round(mbd['active'] / (mbd['mbd'] / 100),0)
```

```python
mbd.head()
```

```python
mbd_pivot = pd.pivot_table(mbd, values='mbd', index=['cfips']
                           , columns = ['first_day_of_month'], aggfunc=np.sum)

avg_pop = pd.pivot_table(mbd, values='pop', index=['cfips']
                           , aggfunc=np.average)
```

```python
mbd_pivot
```

```python
mbd_var = mbd_pivot.var(axis = 'columns', skipna = True, numeric_only = True)
```

```python
mbd_var = pd.DataFrame(mbd_var)
mbd_var.reset_index(inplace=True)
mbd_var.head
```

```python
mbd_var = mbd_var.rename(columns={0:'mbd variance'})
```

```python
mbd_var
```

```python
merged = pd.merge(mbd_var, avg_pop, on="cfips")
```

```python
merged
```

```python
#merged['mbd variance'] = (merged['mbd variance']-merged['mbd variance'].min())/(merged['mbd variance'].max()-merged['mbd variance'].min())

merged['mbd variance'] = np.where(merged['mbd variance'] > 2, 2, merged['mbd variance'])

merged['population bucket'] = np.where(merged['pop'] < 500, '< 500'
                                       , np.where(merged['pop'] < 1000, '< 1000'
                                                  , '> 1000'))
                                       
                                                  
                                        
merged.head()
```

```python
sns.boxplot(x = merged['population bucket'],
            y = merged['mbd variance'],
           order=['< 500', '< 1000', '> 1000'])
```
